```python
def extract_rd_constrants(file_name):
    models = pickle.load(open(file_name,'rb'))
    
    rdes = np.array([[model[i].rd_eta for i in range(len(model))] for model in models])
    rdds = np.array([[model[i].rd_delta for i in range(len(model))] for model in models])
    dneff_cmbs = np.array([[model[i].dneff_cmb for i in range(len(model))] for model in models])
    dneff_bbns = np.array([[model[i].dneff_bbn for i in range(len(model))] for model in models])
    eta_sis = np.array([[model[i].eta_si_per_mass for i in range(len(model))] for model in models])
    deltas_sis = np.array([[model[i].delta_si_per_mass for i in range(len(model))] for model in models])
    sis = (eta_sis * rdes**2 + deltas_sis * rdds**2) / omega_h2_cdm**2
    
    return {
        'rd_eta': rdes,
        'rd_delta': rdds, 
        'dneff_cmb': dneff_cmbs,
        'dneff_bbn': dneff_bbns,
        'si': sis
    }
```

```python
class LaplaceInterp:
    def __init__(self, matrix):
        self.mat = matrix
        self.ii = matrix.shape[0]
        self.jj = matrix.shape[1]

        self.n = self.ii * self.jj
        self.it = 0
        self.b = np.zeros(self.n, dtype=float)
        self.y = np.zeros(self.n, dtype=float)
        self.mask = np.zeros(self.n, dtype=float)

        vl = 0.0
        i = 0
        j = 0
        for k in range(self.n):
            i = int(k / self.jj)
            j = k - i * self.jj
            if self.mat[i][j] < np.nan:
                self.b[k] = self.mat[i][j]
                self.y[k] = self.mat[i][j]
                vl = self.mat[i][j]
                self.mask[k] = 1

            else:
                self.b[k] = 0.
                self.y[k] = vl
                self.mask[k] = 0
                
    def __snrm(self, sx, itol):
        n = sx.shape[0]
        if itol <= 3:
            ans = 0.0
            for i in range(n):
                ans += sx[i]**2
            return np.sqrt(ans)
        else:
            isamax = 0
            for i in range(n):
                if abs(sx[i]) > abs(sx[isamax]):
                    isamax = i
            return abs(sx[isamax])
                
    def __solve(self, b, x, itol, tol, itmax, it, err):
        bkden = 1.0
        EPS = 1.0e-14
        n = b.shape[0]
        p = np.zeros(n, dtype=float)
        pp = np.zeros(n, dtype=float)
        r = np.zeros(n, dtype=float)
        rr = np.zeros(n, dtype=float)
        z = np.zeros(n, dtype=float)
        zz = np.zeros(n, dtype=float)
        self.atimes(x, r, 0)

        for j in range(n):
            r[j] = b[j] - r[j]
            rr[j] = r[j]

        if itol == 1:
            bnrm = self.__snrm(b, itol)
            self.asolve(r, z, 0)
        elif itol == 2:
            self.asolve(b, z, 0)
            bnrm = self.__snrm(z, itol)
            self.asolve(r, z, 0)
        elif itol == 3 or itol == 4:
            self.asolve(b, z, 0)
            bnrm = self.__snrm(z, itol)
            self.asolve(r, z, 0)
            znrm = self.__snrm(z, itol)
        else:
            raise ValueError("Invalid itol in Linbcg.")

        it = 0
        while it < itmax:
            it += 1
            self.asolve(rr, zz, 1)
            bknum = 0.0
            for j in range(n):
                bknum += z[j] * rr[j]
            if it == 1:
                for j in range(n):
                    p[j] = z[j]
                    pp[j] = zz[j]
            else:
                bk = bknum / bkden
                for j in range(n):
                    p[j] = bk * p[j] + z[j]
                    pp[j] = bk * pp[j] + zz[j]
            bkden = bknum
            self.atimes(p, z, 0)
            akden = 0.0
            for j in range(n):
                akden += z[j] * pp[j]
            ak = bknum / akden
            self.atimes(pp, zz, 1)
            for j in range(n):
                x[j] += ak * p[j]
                r[j] -= ak * z[j]
                rr[j] -= ak * zz[j]
            self.asolve(r, z, 0)

            if itol == 1:
                err = self.__snrm(r, itol) / bnrm
            elif itol == 2:
                err = self.__snrm(z, itol) / bnrm
            elif itol == 3 or itol == 4:
                zm1nrm = znrm
                znrm = self.__snrm(z, itol)
                if abs(zm1nrm - znrm) > EPS * znrm:
                    dxnrm = abs(ak) * self.__snrm(p, itol)
                    err = znrm / abs(zm1nrm - znrm) * dxnrm
                else:
                    err = znrm / bnrm
                    continue
                xnrm = self.__snrm(x, itol)
                if err <= 0.5 * xnrm:
                    err /= xnrm
                else:
                    err = znrm / bnrm
                    continue
            if err <= tol:
                break

    def asolve(self, b, x, itrnsp):
        n = b.size()
        for i in range(n):
            x[i] = b[i]

    def atimes(self, x, r, itrnsp):
        n = r.shape[0]
        for k in range(n):
            r[k] = 0.0
        for k in range(n):
            i = k / self.jj
            j = k - i * self.jj
            if self.mask[k]:
                r[k] += x[k]
            elif 0 < i < self.ii - 1 and 0 < j < self.jj - 1:
                if itrnsp:
                    r[k] += x[k]
                    delta = -0.25 * x[k]
                    r[k - 1] += delta
                    r[k + 1] += delta
                    r[k - self.jj] += delta
                    r[k + self.jj] += delta
                else:
                    r[k] = x[k] - 0.25 * (x[k - 1] + x[k + 1] + x[k + self.jj] + x[k - self.jj])
            elif 0 < i < self.ii - 1:
                if itrnsp:
                    r[k] += x[k]
                    delta = -0.5 * x[k]
                    r[k - self.jj] += delta
                    r[k + self.jj] += delta
                else:
                    r[k] = x[k] - 0.5 * (x[k + self.jj] + x[k - self.jj])
            elif 0 < j < self.jj - 1:
                if itrnsp:
                    r[k] += x[k]
                    delta = -0.5 * x[k]
                    r[k - 1] += delta
                    r[k + 1] += delta
                else:
                    r[k] = x[k] - 0.5 * (x[k + 1] + x[k - 1])
            else:
                jjt = self.jj if i == 0 else -self.jj
                it = 1 if j == 0 else -1
                if itrnsp:
                    r[k] += x[k]
                    delta = -0.5 * x[k]
                    r[k + jjt] += delta
                    r[k + it] += delta
                else:
                    r[k] = x[k] - 0.5 * (x[k + jjt] + x[k + it])

    def solve(self, tol=1.e-6, itmax=-1):
        if itmax <= 0:
            itmax = 2 * max(self.ii, self.jj)
        err = 0.0
        self.__solve(self.b, self.y, 1, tol, itmax, self.it, err)
        k = 0
        for i in range(self.ii):
            for j in range(self.jj):
                k += 1
                self.mat[i][j] = self.y[k]
        return err
```
